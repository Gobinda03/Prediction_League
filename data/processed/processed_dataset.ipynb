# Preprocessing the datasets

```python
import pandas as pd
```

```python
# loading the datasets
matches = pd.read_csv("../raw/matches.csv")
deliveries = pd.read_csv("../raw/deliveries.csv")
```

### Matches dataset cleaning

```python
matches.columns
```

```python

matches['date'] = pd.to_datetime(matches['date'])
matches = matches.sort_values('date').reset_index(drop=True)

matches['year'] = matches['date'].dt.year
matches = matches[matches['year'] >= 2015]
```

```python
matches['venue'].nunique()
```

```python
# Team standardization
matches.replace({
    'Delhi Daredevils': 'Delhi Capitals',
    'Kings XI Punjab': 'Punjab Kings',
    'Royal Challengers Bangalore': 'Royal Challengers Bengaluru',
    'Rising Pune Supergiants': 'Rising Pune Supergiant'
}, inplace=True)


# short names
team_short = {
    'Chennai Super Kings': 'CSK',
    'Mumbai Indians': 'MI',
    'Royal Challengers Bengaluru': 'RCB',
    'Kolkata Knight Riders': 'KKR',
    'Delhi Capitals': 'DC',
    'Punjab Kings': 'PBKS',
    'Rajasthan Royals': 'RR',
    'Sunrisers Hyderabad': 'SRH',
    'Lucknow Super Giants': 'LSG',
    'Gujarat Titans': 'GT',
    'Rising Pune Supergiant': 'RPS',
    'Gujarat Lions': 'GL'
}

team_cols = ['team1','team2','toss_winner','winner']

for col in team_cols:
    matches[col] = matches[col].replace(team_short)
```

```python
# Venue 
venue_mapping = {
    # Eden Gardens
    'Eden Gardens, Kolkata': 'Eden Gardens',

    # Chinnaswamy
    'M.Chinnaswamy Stadium': 'M Chinnaswamy Stadium',
    'M Chinnaswamy Stadium, Bengaluru': 'M Chinnaswamy Stadium',

    # Chepauk
    'MA Chidambaram Stadium, Chepauk, Chennai': 'MA Chidambaram Stadium',
    'MA Chidambaram Stadium, Chepauk': 'MA Chidambaram Stadium',

    # Wankhede
    'Wankhede Stadium, Mumbai': 'Wankhede Stadium',

    # Rajiv Gandhi Stadium
    'Rajiv Gandhi International Stadium, Uppal, Hyderabad':
        'Rajiv Gandhi International Stadium',
    'Rajiv Gandhi International Stadium, Uppal':
        'Rajiv Gandhi International Stadium',

    # DY Patil
    'Dr DY Patil Sports Academy, Mumbai': 'Dr DY Patil Sports Academy',

    # Brabourne
    'Brabourne Stadium, Mumbai': 'Brabourne Stadium',

    # Arun Jaitley
    'Arun Jaitley Stadium, Delhi': 'Arun Jaitley Stadium',

    # Punjab Cricket Association
    'Punjab Cricket Association IS Bindra Stadium, Mohali':
        'Punjab Cricket Association Stadium',
    'Punjab Cricket Association IS Bindra Stadium, Mohali, Chandigarh':
        'Punjab Cricket Association Stadium',
    'Punjab Cricket Association IS Bindra Stadium':
        'Punjab Cricket Association Stadium',

    # Sawai Mansingh
    'Sawai Mansingh Stadium, Jaipur': 'Sawai Mansingh Stadium',

    # Maharashtra Cricket Association
    'Maharashtra Cricket Association Stadium, Pune':
        'Maharashtra Cricket Association Stadium',

    # ACA-VDCA
    'Dr. Y.S. Rajasekhara Reddy ACA-VDCA Cricket Stadium, Visakhapatnam':
        'Dr. Y.S. Rajasekhara Reddy ACA-VDCA Cricket Stadium',

    # Narendra Modi
    'Sardar Patel Stadium, Motera': 'Narendra Modi Stadium'
}

matches['venue'] = matches['venue'].replace(venue_mapping)
```

```python
current_teams = ['CSK','MI','RCB','KKR','DC','PBKS','RR','SRH','LSG','GT']

matches = matches[
    matches['team1'].isin(current_teams) &
    matches['team2'].isin(current_teams)
]


# reset index
matches.reset_index(drop=True, inplace=True)


print("✅ IPL Dataset Cleaned Successfully")
print("Teams:", matches['team1'].unique())
print("Total Venues:", matches['venue'].nunique())
```

```python
matches.to_csv('matches_new.csv', index=False)
```

### Deliveries dataset cleaning

```python
deliveries.columns
```

```python
deliveries['batting_team'].unique()
```

```python
# Team standardization
deliveries.replace({
    'Delhi Daredevils': 'Delhi Capitals',
    'Kings XI Punjab': 'Punjab Kings',
    'Royal Challengers Bangalore': 'Royal Challengers Bengaluru',
    'Rising Pune Supergiants': 'Rising Pune Supergiant'
}, inplace=True)


# short
team_short = {
    'Chennai Super Kings': 'CSK',
    'Mumbai Indians': 'MI',
    'Royal Challengers Bengaluru': 'RCB',
    'Kolkata Knight Riders': 'KKR',
    'Delhi Capitals': 'DC',
    'Punjab Kings': 'PBKS',
    'Rajasthan Royals': 'RR',
    'Sunrisers Hyderabad': 'SRH',
    'Lucknow Super Giants': 'LSG',
    'Gujarat Titans': 'GT',
    'Rising Pune Supergiant': 'RPS',
    'Gujarat Lions': 'GL'
}

team_cols = ['batting_team', 'bowling_team']

for team in team_cols:
    deliveries[team] = deliveries[team].replace(team_short)
```

```python
deliveries.to_csv('deliveries_new.csv', index=False)
```
